```python
import torch
import matplotlib.pyplot as plt
import numpy as np

dataset = torch.load("../001.pt")
print(dataset[50]['speed0_data'] - dataset[5]['speed0_data'])
```

```python
all_ep_data = []
for d in range(len(dataset)):
    data = dataset[d]
    all_player_data = []
    for i in range(2):
        x_coord = data['x{}_data'.format(i)]
        y_coord = data['y{}_data'.format(i)]
        heading = data['yaw{}_data'.format(i)]
        vel = data['speed{}_data'.format(i)]
        intent = data['intention_{}'.format(i)]
        player_data = [x_coord, y_coord, heading, vel, intent]
        all_player_data.append(player_data)
    all_ep_data.append(all_player_data)
```

```python
fig_vel, axes_vel = plt.subplots(nrows=2)
for i in range(len(all_ep_data)):
    ep_data = all_ep_data[i]
    if ep_data[0][4] == (3,1) and ep_data[1][4] == (0,2):
        for j in range(2):
            player_data = ep_data[j]
            timestep = np.arange(len(player_data[3]))
            axes_vel[j].plot(timestep, player_data[3])
plt.show()
```
